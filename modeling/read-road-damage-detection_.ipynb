## **Generalized Form for different models**

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetB3, ResNet50, DenseNet121
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt

# -----------------------------
# Function to load datasets
# -----------------------------
def load_datasets(train_dir, val_dir, test_dir, image_size=(224, 224), batch_size=32):
    train_ds = image_dataset_from_directory(
        train_dir,
        image_size=image_size,
        batch_size=batch_size,
        label_mode='binary'
    )
    val_ds = image_dataset_from_directory(
        val_dir,
        image_size=image_size,
        batch_size=batch_size,
        label_mode='binary'
    )
    test_ds = image_dataset_from_directory(
        test_dir,
        image_size=image_size,
        batch_size=batch_size,
        label_mode='binary'
    )

    AUTOTUNE = tf.data.AUTOTUNE
    train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
    val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)
    test_ds = test_ds.prefetch(buffer_size=AUTOTUNE)

    return train_ds, val_ds, test_ds

# -----------------------------
# Function for data augmentation
# -----------------------------
def get_data_augmentation():
    return tf.keras.Sequential([
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.1),
        layers.RandomContrast(0.2),
        layers.RandomBrightness(0.2),
    ])

# -----------------------------
# Function to select model backbone
# -----------------------------
def get_backbone(name):
    if name == "EfficientNetB3":
        return EfficientNetB3, tf.keras.applications.efficientnet.preprocess_input
    elif name == "ResNet50":
        return ResNet50, tf.keras.applications.resnet.preprocess_input
    elif name == "DenseNet121":
        return DenseNet121, tf.keras.applications.densenet.preprocess_input
    else:
        raise ValueError("Invalid model name")

# -----------------------------
# Function to build model
# -----------------------------
def build_model(backbone, preprocess, data_augmentation, input_shape=(224, 224, 3)):
    base_model = backbone(
        include_top=False,
        input_shape=input_shape,
        weights='imagenet'
    )
    base_model.trainable = False

    inputs = layers.Input(shape=input_shape)
    x = data_augmentation(inputs)
    x = preprocess(x)
    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.2)(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model(inputs, outputs)

    return model, base_model

# -----------------------------
# Function to compile model
# -----------------------------
def compile_model(model, learning_rate=1e-3):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

# -----------------------------
# Function to set up callbacks
# -----------------------------
def setup_callbacks(model_name):
    checkpoint_cb = ModelCheckpoint(f"best_{model_name}.keras", save_best_only=True)
    earlystop_cb = EarlyStopping(patience=5, restore_best_weights=True)
    lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7)
    return [checkpoint_cb, earlystop_cb, lr_scheduler]

# -----------------------------
# Function to train model
# -----------------------------
def train_model(model, train_ds, val_ds, epochs=20, callbacks=None):
    return model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=epochs,
        callbacks=callbacks
    )

# -----------------------------
# Function for fine-tuning
# -----------------------------
def fine_tune_model(model, base_model, train_ds, val_ds, epochs=20, callbacks=None, fine_tune_at=None):
    base_model.trainable = True
    for layer in base_model.layers[:fine_tune_at]:
        layer.trainable = False

    compile_model(model, learning_rate=1e-5)
    
    return model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=epochs,
        callbacks=callbacks
    )

# -----------------------------
# Function to evaluate model
# -----------------------------
def evaluate_model(model, test_ds):
    test_loss, test_acc = model.evaluate(test_ds)
    print(f"\n✅ Test Accuracy: {test_acc:.4f} | Loss: {test_loss:.4f}")

# -----------------------------
# Function to save model
# -----------------------------
def save_model(model, model_name):
    model.save(f"{model_name}_final_model.keras")

# -----------------------------
# Function to plot training history
# -----------------------------
def plot_training(histories, labels, title_prefix):
    plt.figure(figsize=(14, 5))

    # Accuracy Plot
    plt.subplot(1, 2, 1)
    for history, label in zip(histories, labels):
        plt.plot(history.history['accuracy'], label=f'{label} Train')
        plt.plot(history.history['val_accuracy'], label=f'{label} Val')
    plt.title(f'{title_prefix} Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss Plot
    plt.subplot(1, 2, 2)
    for history, label in zip(histories, labels):
        plt.plot(history.history['loss'], label=f'{label} Train')
        plt.plot(history.history['val_loss'], label=f'{label} Val')
    plt.title(f'{title_prefix} Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.savefig(f"{model_name}_training_plot.png")
    plt.show()

print("This Cell Done")

def main(model_nam):
    # Set model choice
    model_name = "ResNet50"

    # Load datasets
    train_ds, val_ds, test_ds = load_datasets(
        'binary_dataset/train',
        'binary_dataset/valid',
        'binary_dataset/test'
    )

    # Data Augmentation
    data_augmentation = get_data_augmentation()

    # Select Backbone and Preprocessing Function
    Backbone, preprocess = get_backbone(model_name)

    # Build model
    model, base_model = build_model(Backbone, preprocess, data_augmentation)

    # Compile model
    compile_model(model)

    # Set up callbacks
    callbacks = setup_callbacks(model_name)

    # Train model
    history = train_model(model, train_ds, val_ds, epochs=20, callbacks=callbacks)

    # Fine-tune model
    fine_tune_at = len(base_model.layers) // 2
    history_finetune = fine_tune_model(model, base_model, train_ds, val_ds, epochs=20, callbacks=callbacks, fine_tune_at=fine_tune_at)

    # Evaluate model
    evaluate_model(model, test_ds)

    # Save model
    save_model(model, model_name)

    # Plot training history
    plot_training([history, history_finetune], ["Initial", "Fine-tune"], model_name)

print("This cell done")
```

```python
def main(model_nam):
    # Set model choice
    model_name = "ResNet50"

    # Load datasets
    train_ds, val_ds, test_ds = load_datasets(
        'binary_dataset/train',
        'binary_dataset/valid',
        'binary_dataset/test'
    )

    # Data Augmentation
    data_augmentation = get_data_augmentation()

    # Select Backbone and Preprocessing Function
    Backbone, preprocess = get_backbone(model_name)

    # Build model
    model, base_model = build_model(Backbone, preprocess, data_augmentation)

    # Compile model
    compile_model(model)

    # Set up callbacks
    callbacks = setup_callbacks(model_name)

    # Train model
    history = train_model(model, train_ds, val_ds, epochs=20, callbacks=callbacks)

    # Fine-tune model
    fine_tune_at = len(base_model.layers) // 2
    history_finetune = fine_tune_model(model, base_model, train_ds, val_ds, epochs=20, callbacks=callbacks, fine_tune_at=fine_tune_at)

    # Evaluate model
    evaluate_model(model, test_ds)

    # Save model
    save_model(model, model_name)

    # Plot training history
    plot_training([history, history_finetune], ["Initial", "Fine-tune"], model_name)

print("This cell done")
```
